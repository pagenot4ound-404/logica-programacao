## Tabela Verdade


And


```python
print(10 > 2 and 4 < 6)     # V and V - V
print(10 > 2 and 4 > 6)     # V and F - F
print(10 < 2 and 4 < 6)     # V and F - F
print(10 < 2 and 4 > 6)     # F and F - F
```

Or


```python
print(10 > 2 or 4 < 6)      # V or V - V
print(10 > 2 or 4 > 6)      # V or F - V
print(10 < 2 or 4 < 6)      # V or F - V
print(10 < 2 or 4 > 6)      # F or F - F
```

Not

```python
not 10 > 2     # V - F
```

```python
not 10 < 2     # F - V
```

# Condicional

```python
# If
if 5 < 10:
  print('Verdade')
```

```python
# If/else
num1 = 10
num2 = 5
if num1 > num2:
  print('O num1 é o maior número')
else:
  print('O num2 é o maior número')
```

```python
# If / Elif / Else
num1 = 10
num2 = 10
if num1 > num2:
  print('O num1 é o maior número')
elif num1 < num2:
  print('O num2 é o maior número')
else:
  print('Os dois são iguais')
```

#Condicional com operadores lógicos

```python
num1 = 10
num2 = 5
num3 = 2
if (num1 > num2) and (num1 > num3):
  print('O num1 é o maior número')
else:
  print('O num1 não é o maior número')
```

# Estrutura de Repetição

For

```python
# For Range
for contador in range(10):
    print(contador)
```

```python
# For Range + Break
for contador in range(1000):
    print(contador)
    if contador == 2:
        break
```

```python
lista = ["Teste1", "Teste2"]
```

```python
# For Simples
for val in lista:
    print(val)
```

```python
# For com Enumerate
for i, val in enumerate(lista):
    print(i, val)
```

For com Operadores

```python
lista2 = [2, 3, 4]
```

```python
for val in lista2:
    val += 1
    print(val)
```

```python
for val in lista2:
    val *= 2
    print(val)
```

While

```python
val = 1
while (val < 6):
    val *= 2
    print(val)
```

```python
# While com Else
val = 10
while val < 6:
     val *= 2
else:
     print('Valor de val é:', val)
```
